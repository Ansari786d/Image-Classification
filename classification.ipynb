```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
training_set= train_datagen.flow_from_directory('/content/Section 40 - Convolutional Neural Networks (CNN)/dataset/training_set',
                                                target_size=(64,64),
                                                batch_size=16,
                                                class_mode = 'binary')
```

```python
test_gen = ImageDataGenerator(rescale=1./255)
test_set = test_gen.flow_from_directory('/content/Section 40 - Convolutional Neural Networks (CNN)/dataset/test_set',
                                        target_size = (64,64),
                                        batch_size=16,
                                        class_mode='binary')
```

```python
cnn=tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64,64,3] ))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x=training_set, validation_data=test_set, epochs=25)
```

```python
training_set.class_indices
```

Predicting  new image

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('/content/Section 40 - Convolutional Neural Networks (CNN)/dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'

print(prediction)
```
